```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

df = pd.read_csv("./Hitters.csv")
df = df.dropna()
dms = pd.get_dummies(df[["League", "Division", "NewLeague"]])
y = df["Salary"]
x_ = df.drop(["Salary", "League", "Division", "NewLeague"], axis = 1).astype("float64")
x = pd.concat([x_, dms[["League_N", "Division_W", "NewLeague_N"]]], axis = 1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 99)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor().fit(x_train, y_train)
```

```python
knn_model.n_neighbors
```

```python
pred_y = knn_model.predict(x_test)
```

```python
#RMSE value for the model
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test, pred_y))
```

```python
RMSE = []

for i in range(10):
    knn_model = KNeighborsRegressor(n_neighbors = i + 1).fit(x_train, y_train)
    pred_y = knn_model.predict(x_test)
    RMSE.append(np.sqrt(mean_squared_error(y_test, pred_y)))
```

```python
RMSE
```

```python
#Tuning manually
knn_model = KNeighborsRegressor(n_neighbors = 6).fit(x_train, y_train)
```

```python
#GridSearchCV
knn_params = {"n_neighbors": np.arange(1, 30, 1)}
```

```python
knn = KNeighborsRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
knn_cv_model = GridSearchCV(knn, knn_params, cv = 10).fit(x_train, y_train)
```

```python
knn_cv_model.best_params_
```

```python
knn_tuned = KNeighborsRegressor(n_neighbors = knn_cv_model.best_params_["n_neighbors"]).fit(x_train, y_train)
```

```python
pred_y = knn_tuned.predict(x_test)
np.sqrt(mean_squared_error(y_test, pred_y))
```
